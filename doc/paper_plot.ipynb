## Gamma 函数模拟
scipy 函数库中的Gamma函数
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gamma.html
numpy 函数库中：
https://numpy.org/doc/stable/reference/random/generated/numpy.random.gamma.html
对比numpy，scipy中的gamma函数有ppf、pdf等方便的子类，而numpy只能生成随机数。

```python
from scipy.stats import gamma
a=0.01
b=0.01
scale = 1/b

x = np.linspace(0,20, 100)    
y = gamma.pdf(x, a=a, loc=0, scale=scale)
plt.plot(x, y)
```

## 求解显著角集合事，power peak 方法并不准确

```python
import numpy as np
import math, cmath
import matplotlib as mpl
import matplotlib.pyplot as plt

N = 64
x = np.array(range(N))

e_seq1 = np.exp(1j*2*np.pi*x*0.45)
e_seq2 = np.exp(1j*2*np.pi*x*0.35)


y1 = np.fft.fft(e_seq1)
y2 = np.fft.fft(e_seq2)
y3 = np.fft.fft(e_seq1+e_seq2)

fig, ax = plt.subplots() # 创建图实例
ax.scatter(x, np.abs(y1), label='e_seq1') # 作y1 = x 图，并标记此线名为linear

ax.scatter(x, np.abs(y2), label='e_seq2') #作y2 = x^2 图，并标记此线名为quadratic

ax.plot(x, np.abs(y3), label='e_seq1+e_seq2',color='g') # 作y3 = x^3 图，并标记此线名为cubic
ax.plot(x, np.abs(y1+y2), label='fft(e_seq1)+fft(e_seq2)',color='b') # 作y3 = x^3 图，并标记此线名为cubic
ax.legend() #自动检测要在图例中显示的元素，并且显示
```

# DFT Matrix

```python
import numpy as np
import math, cmath
import matplotlib as mpl
import matplotlib.pyplot as plt

N = 64
x = np.array(range(N))

# e_seq1 = np.mat()
e_seq1 = np.mat(np.reshape(np.exp(1j*2*np.pi*x*0.21), (-1,1)))# print(e_seq1)
e_seq2 = np.exp(1j*2*np.pi*x*0.35)


dftmtx = np.mat(np.fft.fft(np.eye(N)))
plt.plot(np.array(range(N)), np.abs(np.array(dftmtx * e_seq1)))

def mydiag(matrix):
    return np.mat(np.diag(np.array(matrix).reshape(1,-1)[0]))

distort_dftmat = mydiag(e_seq1) * dftmtx

# Z is the 2-D data matrix
    # Z = np.abs(dftmtx.H*dftmtx)
def my2Dplot(Z):
    plt.plot(np.array(range(N)), np.abs(np.array(dftmtx * e_seq1)))
    fig, ax = plt.subplots()

    im = ax.imshow(Z, cmap=mpl.cm.RdBu, vmin=abs(Z).min(),vmax=abs(Z)
    .max(), extent=[0, 1, 0, 1])
    im.set_interpolation('bilinear')

    cb = fig.colorbar(im, ax=ax)
    return cb
# my2Dplot(np.abs(dftmtx.H*dftmtx))
my2Dplot(np.abs(distort_dftmat.H*distort_dftmat))
```
